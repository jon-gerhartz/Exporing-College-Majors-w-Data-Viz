# Exploring Initial Questions with Scatter Plots

```python
ax1=recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax1.set_title('Median Salary vs Sample Size')
```

This scatter plot explores the relationship between major popularity and median income. We a week correlation between the two variables. Majors with a similar sample size have a wide range of median salaries. Additionally, majors with different sample sizes have the same salaries. This suggest students are not choosing majors based on expected salary. 

```python
ax2=recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax2.set_title('Unemployment Rate vs Sample Size')
```

There is a week relationship between unemployment rate and sample size. Low sample sizes have a large range of values for unemployment rate. Higher sample sizes have a smaller range of values but this might be because there are not enough values with a large sample size to represent the same variation as the smaller sample sizes. 

```python
ax3=recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax3.set_title('Median Salary vs Full Time')
```

There is a weak nagative relationship between the number of full time employees and the median salary. As the number of full time employees increases, the median salary slightly decreases. There is a large variation for the majors with a small amount of full time workers. Variation decreases as the number of full time workers increases. 

```python
ax4=recent_grads.plot(x='ShareWomen', y='Unemployment_rate',kind='scatter')
ax4.set_title('Unemployment Rate vs % of Women')
```

There is no obvious relationship that can be observed from the visualization of the share of women in a major and the unemployment rate of that major. We can observe that most of the unemployment rates lie between .025 and .125.

```python
ax5=recent_grads.plot(x='Men', y='Median',kind='scatter')
ax5.set_title('Median Salary vs Men')
```

There is no obvious relationship between the number of men in the major and the median salary. 

```python
ax6=recent_grads.plot(x='Women',y='Median', kind='scatter')
ax6.set_title('Median Salary vs Women')
```

There is no obvious relationshp between the number of women in a major and the median salary.

## Summary of intial question visualizations 

A common problem with this initial visualization is that many of the values for sample size, the number of full time workers, and the number of men and women are much smaller than the range represented in the visualization. Because of the large range of the visualization, it is hard to see how small differences in the number of men, women, or full time employees in a major affect the median salary. If we take a small portions of the distribution that is closer to the mean we might be able to observe stronger relationships between these variables and median salary. A few outliers are making these visualizations less useful.

# Exploring Data with Bar Plots

## Share of Women in top ranking majors verses share of women in low ranking majors

```python
women_first=recent_grads.head(10).plot.bar(x='Major',y='ShareWomen')
women_last=recent_grads.tail(10).plot.bar(x='Major',y='ShareWomen')
```

Based on these two charts we see that there is a much larger share of women in the 10 lowest ranking majors than there are in the 10 highest ranking majors. This may indicate a larger pattern in the data that there are a larger share of women in the lower ranking or lower salary majors. We can look at a scatter plot of rank and women to see if this pattern exists throughout the dataset. 

```python
women_plot=recent_grads.plot(x='Rank',y='ShareWomen',kind='scatter')
women_plot.set_title('Major Rank vs Share of Women in Major')
```

After running a scatter plot, we see that there is a positive correlation between rank and the share of women. As rank increases(or median salary decreases) the share of women in that major also increases. Even though we observed above that 55% of the majors were dominated by women, we could infer that most of those majors were lower paying jobs.
